```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
apo_mean_acc = np.array([67.4625, 74.8875, 83.86666667, 86.275, 87.5375])
apo_std_acc = np.array([0.555840205, 0.625, 0.880814017, 0.963068014, 0.385951206])
rand_mean_acc = np.array([66.275, 72.05, 78.9625, 83.6375, 84.9])
rand_std_acc = np.array([0.992051746, 0.940744386, 0.694472222, 0.759797122, 0.63900965])
x = np.array([500, 1000, 2000, 3000, 4000])
```

```python
plt.plot(x, apo_mean_acc, label='APO (Ours)', markersize=8, color='blue', marker='o')
plt.fill_between(x, apo_mean_acc - apo_std_acc, apo_mean_acc + apo_std_acc, alpha=0.4, color='blue')
plt.plot(x, rand_mean_acc, label='Random', markersize=8, color='orange', marker='o')
plt.fill_between(x, rand_mean_acc - rand_std_acc, rand_mean_acc + rand_std_acc, alpha=0.4, color='orange')
plt.legend(fontsize=13)
plt.grid()
plt.xlabel('Total Number of Samples', fontsize=13)
plt.ylabel('Reward Accuracy', fontsize=13)
plt.title('Gemma : Reward Accuracy vs Train Samples (Small Epochs)', fontsize=14)
plt.savefig('./results/Gemma-Reward-Accuracy-Small-Epochs', dpi=800)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_apo = pd.read_csv('./output/Anthropic_small_Gemma_apo_le_reward.csv')
df_rand = pd.read_csv('./output/Anthropic_small_Gemma_random_le_reward.csv')
df_apo_2 = pd.read_csv('./output/Anthropic_small_Gemma_apo_le_reward_2.csv')
df_rand_2 = pd.read_csv('./output/Anthropic_small_Gemma_random_le_reward_2.csv')
df_chosen = pd.read_csv('./output/Anthropic_small_Gemma_chosen_le_reward.csv')
df_chosen_2 = pd.read_csv('./output/Anthropic_small_Gemma_chosen_le_reward_2.csv')
```

```python
reward_apo = np.array(pd.concat([df_apo['reward'], df_apo_2['reward']]))
reward_rand = np.array(pd.concat([df_rand['reward'], df_rand_2['reward']]))
reward_chosen = np.array(pd.concat([df_chosen['reward'], df_chosen_2['reward']]))
```

```python
plt.hist(reward_apo, alpha=0.6, label='APO (Ours)', bins=50)
plt.hist(reward_rand, alpha=0.6, label='Random', bins=50)
plt.legend()
plt.show()
```

```python
sns.kdeplot(reward_rand, color='orange', label='Random', fill=True, alpha=0.45)
sns.kdeplot(reward_apo, color='b', label='APO (Ours)', fill=True, alpha=0.45)
plt.xlabel('Rewards')
plt.ylabel('Density')
plt.grid()
plt.show()
```

```python
np.mean(reward_apo > reward_chosen) * 100
```
